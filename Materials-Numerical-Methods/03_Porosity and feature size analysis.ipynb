```python
#怕有版權問題沒使用真實圖片做分析
import numpy as np
import matplotlib.pyplot as plt

# 1. 模擬一張顯微鏡影像 (Simulate Microscopy Image)
# 建立一個 100x100 的矩陣，代表材料基質
image_size = 100
image = np.ones((image_size, image_size))

# 隨機產生一些「孔隙」 (Pores)
np.random.seed(42)
for _ in range(15):
    y, x = np.random.randint(0, 100, size=2)
    radius = np.random.randint(3, 8)
    # 在矩陣中劃出圓形的孔隙
    Y, X = np.ogrid[:image_size, :image_size]
    mask = (X - x)**2 + (Y - y)**2 <= radius**2
    image[mask] = 0  # 將孔隙部分設為 0 (黑色)

# 2. 計算孔隙率 (Calculate Porosity)
# 統計 0 (孔隙) 的數量
total_pixels = image.size
pore_pixels = np.count_nonzero(image == 0)
porosity = (pore_pixels / total_pixels) * 100

# 3. 影像處理：找出孔隙的邊界 (Edge Detection logic using NumPy)
# 這裡簡單利用鄰近像素差異來標示邊界
shifted_image = np.roll(image, shift=1, axis=0)
edges = np.abs(image - shifted_image)

# 4. 繪圖與呈現 (Visualization)
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# 左圖：原始顯微影像
ax[0].imshow(image, cmap='gray')
ax[0].set_title(f"Microscopy Image\nCalculated Porosity: {porosity:.2f}%")
ax[0].axis('off')

# 右圖：特徵提取 (邊界偵測)
ax[1].imshow(edges, cmap='hot')
ax[1].set_title("Pore Edge Detection")
ax[1].axis('off')

plt.tight_layout()
plt.show()

print(f"Analysis Report:")
print(f"Total Area: {total_pixels} pixels")
print(f"Pore Area: {pore_pixels} pixels")
print(f"Porosity: {porosity:.2f}%")
```
