```python
import pandas as pd
import numpy as np

#讓 Python 畫圖
import matplotlib.pyplot as plt

# 模擬材料拉伸數據 (Strain, Stress)
data = {
    'Strain': [0.000, 0.001, 0.002, 0.003, 0.004, 0.005, 0.007, 0.010],
    'Stress': [0, 210, 420, 630, 840, 950, 1050, 1100] # unit: MPa
}

#轉成表格^^
df = pd.DataFrame(data)

# 1. 篩選彈性區間 (假設在應變 0.004 以下為線性)
elastic_df = df[df['Strain'] <= 0.004]

# 2. 線性擬合 (一階多項式 y = ax + b)
# E 是斜率，intercept 是截距
E, intercept = np.polyfit(elastic_df['Strain'], elastic_df['Stress'], 1)

print(f"Calculated Young's modulus (E): {E/1000:.2f} GPa")
```
